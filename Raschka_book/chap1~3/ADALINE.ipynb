```python
import ADALINE
import os 
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data', header=None, encoding='utf-8')
```

```python
df.tail()
```

```python
y = df.iloc[0:100, 4].values
```

```python
y = np.where(y == 'Iris-setosa', 0, 1)
```

```python
X = df.iloc[0:100, [0, 2]].values
```

```python
ada1 = ADALINE.ADALINE_full_batch(n_iter=15, eta=0.1).fit(X, y)
```

```python
ada1.plot_losses(logplot=1)
```

```python
ada2 = ADALINE.ADALINE_full_batch(n_iter=15, eta=0.0001).fit(X, y)
```

```python
ada2.plot_losses()
```

```python
X_std = np.copy(X)
```

```python
X_std[:, 0] = (X[:, 0] - X[:, 0].mean()) / X[:, 0].std() 
```

```python
X_std[:, 1] = (X[:, 1] - X[:, 1].mean()) / X[:, 1].std()
```

```python
ada_sgd = ADALINE.ADALINE_sgd(n_iter=15, eta=0.01, random_state=1) 
```

```python
ada_sgd.fit(X_std, y)
```

```python
ada_sgd.plot_decision_regions(X_std, y)
```

```python
ada_sgd.plot_losses()
```
